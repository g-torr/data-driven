Data are shifted so protocol is the same in all of them. The trial changes at different times, therefore I have to cut the signal to have the same lenght. Then I averaged out. Difference on trial, i.e. the time between mental task is the same (apart of changes of 1 frame, which are neglected). Therefore I select the last trial that is the one with the shortest wait before the protocol begins.

```python
import mat2python
data,trial,classes=mat2python.mat2python(patient=1,repetition_number=0)
oxy=data[trial[0]-219:trial[0]-219+2642,:52]
for pat in arange(2, 9):
    data,trial,classes=mat2python.mat2python(patient=pat,repetition_number=0)
    #sampling rate 10 Hz
    oxy+=data[trial[0]-219:trial[0]-219+2642,:52] #219 is when the protocol starts in the 8th patient.
oxy=oxy/8
oxy=data[:,0:52]
roi1=[46,47,48]
roi2=[18,28,29]
roi3=[13, 23, 24]
signal=oxy[:,roi1+roi2+roi3]
```

Correlation coefficient for the covariance of the averaged signal. Given the channel  $i,j$
$$
\frac{cov(x_i,x_j)}{\sigma_i\sigma_j}
$$

```python
figure(figsize=(4,3))
imshow(corrcoef(oxy[:,roi1+roi2+roi3],rowvar=False),aspect="equal",cmap="inferno")
print(shape(oxy))
ax=gca()
ax.set_xticks(arange(len(roi1+roi2+roi3)))
ax.set_yticks(arange(len(roi1+roi2+roi3)))
xlabel("channel",fontsize=14)
ylabel("channel",fontsize=14)
ax.set_xticklabels(list(array(roi1)+1)+list(array(roi2)+1)+list(array(roi3)+1))
ax.set_yticklabels(list(array(roi1)+1)+list(array(roi2)+1)+list(array(roi3)+1))
cbar=colorbar(orientation="vertical")
cbar.set_label('Correlation coefficient',fontsize=14)
tight_layout()
#savefig("./figure/averaged-covariance.pdf")
```

## Multiple changepoint detection

### Optimal decision

```python
# detection
algo = rpt.Dynp(model="mahalanobis",min_size=10).fit(oxy[:,roi1+roi2+roi3]) #equivalent to rpt.Dynp(model="mahalanobis", params={"metric": inv(M)})
M_result = algo.predict(n_bkps=12)
rpt.display(oxy[:,roi1], trial,M_result)
savetxt("avareged-optimal_mahalanobis.dat",M_result)
```

### Window based changepoint

```python
# change point detection
model="mahalanobis" # "l1", "rbf", "linear", "normal", "ar"
algo = rpt.Window(width=40, model=model).fit(signal)
my_bkps = algo.predict(n_bkps=12)

# show results
rpt.show.display(signal,trial, my_bkps, figsize=(10, 6))
plt.show()
```

```python
figure(figsize=(4,3))
peak_inds_arr = np.take(algo.inds,arange(0,len(algo.score)))# rupture maps in a misterius way the time. This map convers the internal mapping to frame. 
plot(peak_inds_arr/10,algo.score)
ymin,ymax=gca().get_ylim()
vlines(array(my_bkps)/10,ymin,ymax,linestyle="--",lw=0.8)
xlabel("time (s)",fontsize=14)
ylabel("discrepacy",fontsize=14)

tight_layout()
#savefig("./figure/averaged-discrepancy.pdf")
```
